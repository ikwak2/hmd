```python

def get_features(patient_files_trn) :
    features = dict()
    features['id'] = []
    features['age'] = []
    features['sex'] = []
#     features['hw'] = []
    features['preg'] = []
    features['loc'] = []
    features['mel1'] = []
    features['stft1'] = []
    features['cqt1'] = []
    labels = []

    age_classes = ['Neonate', 'Infant', 'Child', 'Adolescent', 'Young Adult']
    recording_locations = ['AV', 'MV', 'PV', 'TV', 'PhC']

    num_patient_files = len(patient_files_trn)

    for i in range(num_patient_files):

        # Load the current patient data and recordings.
        current_patient_data = load_patient_data(patient_files_trn[i])
        num_locations = get_num_locations(current_patient_data)
        recording_information = current_patient_data.split('\n')[1:num_locations+1]
        for j in range(num_locations) :
            entries = recording_information[j].split(' ')
            recording_file = entries[2]
            filename = os.path.join(data_folder, recording_file)

            # Extract id
            id1 = recording_file.split('_')[0]
            features['id'].append(id1)

            # Extract melspec
            mel1 = feature_extract_melspec(filename)[0]
            features['mel1'].append(mel1)
            
            
            # Extract stft
            stft1 = feature_extract_stft(filename)[0]
            features['stft1'].append(stft1)
            
            
            # Extract cqt
            cqt1 = feature_extract_cqt(filename)[0]
            features['cqt1'].append(cqt1)
            
            
            # Extract age_group
            age_group = get_age(current_patient_data)
            current_age_group = np.zeros(6, dtype=int)
            if age_group in age_classes:
                j = age_classes.index(age_group)
                current_age_group[j] = 1
            else :
                current_age_group[5] = 1
            features['age'].append(current_age_group)

            # Extract sex
            sex = get_sex(current_patient_data)
            sex_features = np.zeros(2, dtype=int)
            if compare_strings(sex, 'Female'):
                sex_features[0] = 1
            elif compare_strings(sex, 'Male'):
                sex_features[1] = 1
            features['sex'].append(sex_features)

            # Extract height and weight.
#             height = get_height(current_patient_data)
#             weight = get_weight(current_patient_data)
#             features['hw'].append(np.array([height, weight]))

            # Extract pregnancy
            is_pregnant = get_pregnancy_status(current_patient_data)
            features['preg'].append(is_pregnant)

            # Extract location
            locations = entries[0]
            num_recording_locations = len(recording_locations)
            loc_features = np.zeros(num_recording_locations)
            if locations in recording_locations:
                j = recording_locations.index(locations)
                loc_features[j] = 1
            features['loc'].append(loc_features)

            # Extract labels and use one-hot encoding.
            current_labels = np.zeros(num_classes, dtype=int)
            label = get_label(current_patient_data)
            if label in classes:
                j = classes.index(label)
                current_labels[j] = 1
            labels.append(current_labels)
        
    M, N = features['mel1'][i].shape
    Y, Z = features['stft1'][i].shape
    X, V = features['cqt1'][i].shape
    
    
    for i in range(len(features['mel1'])) :
        features['mel1'][i] = features['mel1'][i].reshape(M,N,1)

    for i in range(len(features['stft1'])) :
        features['stft1'][i] = features['stft1'][i].reshape(Y,Z,1)
        
    for i in range(len(features['cqt1'])) :
        features['cqt1'][i] = features['cqt1'][i].reshape(X,V,1)
        
        
    labels = np.array(labels)
    
    for k1 in features.keys() :
        features[k1] = np.array(features[k1])
    
    return features, labels
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
age = keras.Input(shape=(6,), name = 'age_cat')
sex = keras.Input(shape=(2,), name = 'sex_cat')
# hw = keras.Input(shape=(2,), name = 'height_weight')
preg = keras.Input(shape=(1,), name = 'is_preg')
loc = keras.Input(shape=(5,), name = 'loc')
mel1 = keras.Input(shape=((100, 313, 1)), name = 'mel')
stft1 = keras.Input(shape=((257, 313, 1)), name = 'stft')
cqt1 = keras.Input(shape=((80, 157, 1)), name = 'cqt')
```

```python
def convolutional_block(x, filter):
    # copy tensor to variable called x_skip
    x_skip = x
    # Layer 1
    x = tf.keras.layers.Conv2D(filter, (3,3), padding = 'same', strides = (2,2))(x)
    x = tf.keras.layers.BatchNormalization(axis=3)(x)
    x = tf.keras.layers.Activation('relu')(x)
    # Layer 2
    x = tf.keras.layers.Conv2D(filter, (3,3), padding = 'same')(x)
    x = tf.keras.layers.BatchNormalization(axis=3)(x)
    # Processing Residue with conv(1,1)
    x_skip = tf.keras.layers.Conv2D(filter, (1,1), strides = (2,2))(x_skip)
    # Add Residue
    x = tf.keras.layers.Add()([x, x_skip])     
    x = tf.keras.layers.Activation('relu')(x)
    return x
```

```python
## age embeddig
age1 = layers.Dense(2, activation = 'relu')(age)

## sex embedding
sex1 = layers.Dense(1, activation = 'relu')(sex)

## hw embedding
# hw1 = layers.Dense(1, activation = 'relu')(hw)

## loc embedding
loc1 = layers.Dense(3, activation = 'relu')(loc)
```

```python
mel2 = convolutional_block(mel1, 16)
mel2 = tf.keras.layers.Dropout(.5)(mel2)
mel2 = convolutional_block(mel2, 32)
mel2 = tf.keras.layers.Dropout(.5)(mel2)
mel2 = convolutional_block(mel2, 64)
mel2 = tf.keras.layers.Dropout(.5)(mel2)
mel2 = convolutional_block(mel2, 128)
mel2 = tf.keras.layers.Dropout(.5)(mel2)
mel2 = convolutional_block(mel2, 256)
mel2 = tf.keras.layers.Dropout(.5)(mel2)
mel2 = layers.GlobalAveragePooling2D()(mel2)


stft2 = convolutional_block(stft1, 16)
stft2 = tf.keras.layers.Dropout(.5)(stft2)
stft2 = convolutional_block(stft2, 32)
stft2 = tf.keras.layers.Dropout(.5)(stft2)
stft2 = convolutional_block(stft2, 64)
stft2 = tf.keras.layers.Dropout(.5)(stft2)
stft2 = convolutional_block(stft2, 128)
stft2 = tf.keras.layers.Dropout(.5)(stft2)
stft2 = convolutional_block(stft2, 256)
stft2 = tf.keras.layers.Dropout(.5)(stft2)
stft2 = layers.GlobalAveragePooling2D()(stft2)

cqt2 = convolutional_block(cqt1, 16)
cqt2 = tf.keras.layers.Dropout(.5)(cqt2)
cqt2 = convolutional_block(cqt2, 32)
cqt2 = tf.keras.layers.Dropout(.5)(cqt2)
cqt2 = convolutional_block(cqt2, 64)
cqt2 = tf.keras.layers.Dropout(.5)(cqt2)
cqt2 = convolutional_block(cqt2, 128)
cqt2 = tf.keras.layers.Dropout(.5)(cqt2)
cqt2 = convolutional_block(cqt2, 256)
cqt2 = tf.keras.layers.Dropout(.5)(cqt2)
cqt2 = layers.GlobalAveragePooling2D()(cqt2)

# mel2 = layers.Conv2D(16, (2,2), activation = 'relu')(mel1)
# mel2 = tf.keras.layers.BatchNormalization()(mel2)
# mel2 = tf.keras.layers.Activation('relu')(mel2)
# mel2 = layers.Conv2D(32, (2,2), activation = 'relu')(mel2)
# mel2 = tf.keras.layers.BatchNormalization()(mel2)
# mel2 = tf.keras.layers.Activation('relu')(mel2)
# mel2 = tf.keras.layers.Dropout(.5)(mel2)
# mel2 = convolutional_block(mel2, 64)
# mel2 = convolutional_block(mel2, 128)
# mel2 = tf.keras.layers.Dropout(.5)(mel2)
# mel2 = convolutional_block(mel2, 256)
# mel2 = layers.GlobalAveragePooling2D()(mel2)


# stft2 = layers.Conv2D(16, (2,2), activation = 'relu')(stft1)
# stft2 = tf.keras.layers.BatchNormalization()(stft2)
# stft2 = tf.keras.layers.Activation('relu')(stft2)
# stft2 = layers.Conv2D(32, (2,2), activation = 'relu')(stft2)
# stft2 = tf.keras.layers.BatchNormalization()(stft2)
# stft2 = tf.keras.layers.Activation('relu')(stft2)
# stft2 = tf.keras.layers.Dropout(.5)(stft2)
# stft2 = convolutional_block(stft2, 64)
# stft2 = convolutional_block(stft2, 128)
# stft2 = tf.keras.layers.Dropout(.5)(stft2)
# stft2 = convolutional_block(stft2, 256)
# stft2 = layers.GlobalAveragePooling2D()(stft2)


# cqt2 = layers.Conv2D(16, (2,2), activation = 'relu')(cqt1)
# cqt2 = tf.keras.layers.BatchNormalization()(cqt2)
# cqt2 = tf.keras.layers.Activation('relu')(cqt2)
# cqt2 = layers.Conv2D(32, (2,2), activation = 'relu')(cqt2)
# cqt2 = tf.keras.layers.BatchNormalization()(cqt2)
# cqt2 = tf.keras.layers.Activation('relu')(cqt2)
# cqt2 = tf.keras.layers.Dropout(.5)(cqt2)
# cqt2 = convolutional_block(cqt2, 64)
# cqt2 = convolutional_block(cqt2, 128)
# cqt2 = tf.keras.layers.Dropout(.5)(cqt2)
# cqt2 = convolutional_block(cqt2, 256)
# cqt2 = layers.GlobalAveragePooling2D()(cqt2)
```

```python
concat2 = layers.Concatenate()([mel2,stft2,cqt2,age1,sex1,loc1,preg])
concat2 = layers.Dense(10, activation = 'relu')(concat2)
concat2 = layers.Dense(3, activation = "softmax")(concat2)
```

```python
model1 = keras.Model(inputs = [age,sex,preg,loc,mel1,stft1,cqt1] , outputs = concat2 )
```

```python
model1.compile(optimizer = "adam", 
             loss = "categorical_crossentropy",
             metrics = "accuracy")
```

```python
import pickle as pk
def save_challenge_model2(model_folder, model, classes, m_name, mel_shape = (100, 313, 1),stft_shape = (257, 313, 1),
                         cqt_shape = (80, 157, 1)):
    os.makedirs(model_folder, exist_ok=True)
    info_fnm = os.path.join(model_folder, 'desc.pk')
    filename = os.path.join(model_folder, m_name + '_model.hdf5')
    model.save(filename)
    d = {'model': m_name, 'classes': classes, 'mel_shape': mel_shape, 'stft_shpae' : stft_shape,
         'cqt_shape' : cqt_shape,'model_fnm': filename}    
    with open(info_fnm, 'wb') as f:
        pk.dump(d, f, pk.HIGHEST_PROTOCOL)
    return 1
    
```

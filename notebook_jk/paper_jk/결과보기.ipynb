```python


data = []
allkeys = []
all_file = glob.glob(model_save_path + '/**')
print(len(all_file))
for i in range(len(all_file)) :
    with open(all_file[i], 'rb') as f:
        data1 = pickle.load(f)
    data.append(data1)
    allkeys += data1.keys()

allkeys = list(set(allkeys))

newlist = {}
for i in range(len(data)) :
    if i == 0 :
        for k1 in allkeys :
            try :
                newlist[k1] = [ data[i][k1] ]
            except :
                newlist[k1] = [-999]
    else :
        for k1 in allkeys :
            try :
                newlist[k1].append(data[i][k1])
            except :
                newlist[k1].append(-999)
                
df1 = pd.DataFrame(newlist)
```
